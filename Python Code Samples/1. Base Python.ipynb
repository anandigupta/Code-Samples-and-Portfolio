### Question 1

(a) What is meant by “reproducibility” and why is it important?

Reproducibility is the ability to replicate the original results from your code by following a specific set of steps with the same dataset. To make work reproducible, code should be readable (concise and clear), portable (easy to move between computers), well-named, repeatable (every step of project should be done in code), and follow good version control. Reproducibility is important in order to verify that the methods and results of research is sound. 


(b) How does version control using Git work?

Git is a version control system that records specific changes to a file or set of files that you can go back to a specific version (basically a snapshot based version control system). You first create a working directory with original files. You can then add in modified files that you want to remember to your staging area. This helps you keep track of certain changes you make at a specific period in time. You then commit the changes, which freezes the changes (creating a snapshot) in that point of time. However, you will be able to "check out" older versions of the working directory and move between versions. 


(c) What is the difference between the is, in, and == operators in Python?

The is operator checks whether both variables point to the same object (saved in the same location in memory) or not. The == operator checks whether the values of the objects pointed to by the two variables have equal values. The in function checks whether a variable is in the specified sequence.  


(d) What is a shallow copy and when do shallow copies occur when programming in Python?

A shallow copy is a copy where changes in one object are reflected in the other. When programming in python, we tend to make copies for objects containing mutable items, so that changes in one copy are not reflected in the other. However, when working with compound objects such as a nested list, the copy function creates a new compound object and inserts references to objects in the original compound object. Thus, changes to objects in the copy will be reflected in the original (shallow copy). To avoid this, we can use the deepcopy function instead. 


(e) When analyzing the runtime using Big-O notation, what part of the algorithm do we care
the most about?

Big-O notation is used to describe how fast a function grows or declines. We care the most about the worst case, i.e the part of the algorithm that grows the fastest (dominant function or order of magnitude function), as this is the maximum number of operations that might be performed for a given problem size.

```python
converted_dict = {}

def country_values(data):
    
    """
    Converts data to long-format dictionary with tuple key
    
    Arguments
    ---------
    data: str 
        A text string indicating the relevant nested list
    
    return
    ------
    A dictionary with country year combinations as the key
    """
#loop through the rows of the list
    for i in range(1,len(data)):
        
#loop through the columns of the list
        for j in range(1, len(data[0])):

#identify the relevant indices needed for the new key and value
            country = data[i][0]
            year = data[0][j]
            value = (data[i][j])
            
#create new dictionary with tuple key and value
            if not value:
                converted_dict[(country, year)] = None
            else:
                converted_dict[(country, year)] = float(value)

    return converted_dict

country_values(country_data)
        
```

The big-O notation for this function is $O(N \cdot M)$ where N is the number of rows and M is the number of columns in my nested list, as the two dominant functions are nested for loops of different sizes.

The big-O notation for this function is $O(N)$ where N is the number of rows in the dictionary, which I loop through in my function.
